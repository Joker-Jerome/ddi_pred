```r
map_df$rxcui <- as.character(map_df$rxcui)
```

```r
mapping_db <- function(df, map_df) {
    map_df_a <- map_df
    map_df_b <- map_df
    colnames(map_df_a)[3] <- "drug_1_drugbank_id"
    colnames(map_df_b)[3] <- "drug_2_drugbank_id"
    tmp_df <- df %>%
        inner_join(map_df_a, by = c("drug_1_rxnorn_id" = "rxcui")) %>%
        inner_join(map_df_b, by = c("drug_2_rxnorm_id" = "rxcui"))
    return(tmp_df)
}
```

```r
liver_ddi_comp <- mapping_db(liver_ddi, map_df)
```

```r
head(liver_ddi_comp)
```

```r
renal_ddi_comp <- mapping_db(renal_ddi, map_df)
heart_ddi_comp <- mapping_db(heart_ddi, map_df)
neuro_ddi_comp <- mapping_db(neuro_ddi, map_df)
```

```r
write.csv(liver_ddi_comp, file = "liver_ddi.csv")
write.csv(renal_ddi_comp, file = "renal_ddi.csv")
write.csv(heart_ddi_comp, file = "heart_ddi.csv")
write.csv(neuro_ddi_comp, file = "neuro_ddi.csv")
```
